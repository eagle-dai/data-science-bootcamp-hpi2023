# Linear Regression

- ### Is a Supervised Learning Technique used in Machine Learning
- ### Main task is to determine the relationship between an independent and dependent variables in which both are continious
- ### In other words it is used for predicting a continious quantity

----
\
### Simply Speaking, Linear Regression is just a fancy name to saying fitting line to some data using least squares

# What does that look like?

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate random data with positive correlation
x_values = np.random.rand(70)
y_values = x_values + np.random.rand(70)*0.5

# Plot the data with a scatter plot
plt.scatter(x_values, y_values)

# Add labels and title to the plot
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Data Example')

# Show the plot
plt.show()
```

# Which line fits the data better?

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import linregress

# Plot the data with a scatter plot
plt.scatter(x_values, y_values)

# Drawing "Regression" lines
x = [0, 1]
y1 = [0.21, 1.3]
y2 = [0.05, 1.1]
y3 = [0.1, 1.5]

# Plot the line
plt.plot(x, y1, '--g')
plt.plot(x, y2, '--b')
plt.plot(x, y3, '--r')

# Add labels and title to the plot
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Data Example')

# Show the plot
plt.show()
```

# Linear Regression using scikit-learn Library

[Linear Regression Sklearn Documentation](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html)

```python
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Reshape the data into 2D arrays
X = x_values.reshape(-1, 1)
y = y_values.reshape(-1, 1)

# Create a LinearRegression object
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)

# Generate predicted values of y
y_pred = model.predict(X)

# Plot the data points and the regression line
plt.scatter(x_values, y_values)
plt.plot(x_values, y_pred, color='green')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Linear Regression Example')

# Show the plot
plt.show()

```

# Least Squares

```python
plt.scatter(x_values, y_values)
plt.plot(x_values, y_pred, color='green')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Linear Regression Example')


for i in range(len(X)):
    plt.plot([X[i], X[i]], [y[i], y_pred[i]], 'r--', alpha=0.5)
```

# How to know how good your model was?

- You can use sklearns function ***score***

- Score gives you he coefficient of determination, or R^2.
- It is a measure that provides information about the goodness of fit of a model.
- In the context of regression it is a statistical measure of how well the regression line approximates the actual data.

```python
print(f"Your Linear Regression model achieved a score of: {model.score(X, y)}\n")
```

# One Very Important remark when deploying any machine learning!

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Plot the data points and the regression line
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.scatter(X_test, y_test, color='green', label='Testing Data')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Data Train-Test Split')

# Add a legend to the plot
plt.legend()

# Show the plot
plt.show()
```

### Continue as before

```python
# Create a LinearRegression object
model = LinearRegression()

# Fit the model to the data
model.fit(X_train, y_train)
print(f"Your Linear Regression model achieved a score of: {model.score(X_test, y_test)}\n")

# Generate predicted values of y
y_pred = model.predict(X_test)

# Plot the data points and the regression line
plt.scatter(X_test, y_test)
plt.plot(X_test, y_pred, color='red')
plt.xlabel('X Test values')
plt.ylabel('Y Test values')
plt.title('Linear Regression Model Prediction')

# Show the plot
plt.show()
```
