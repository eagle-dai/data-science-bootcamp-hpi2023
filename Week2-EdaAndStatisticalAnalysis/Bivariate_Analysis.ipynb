# **Bivariate Analysis**

*  #### This type of data involves two different variables

*  #### The analysis of this type of data deals with causes and relationships

*  ####  The analysis is done to find out the relationship among the two variables



### **Bivariate analysis can be used in a variety of contexts, such as:**

1. #### **Exploring the relationship between two variables:** Bivariate analysis can be used to examine the relationship between two variables, to identify any patterns or trends in the data, and to determine whether the relationship is statistically significant.

2. #### **Comparing groups:** Bivariate analysis can be used to compare the values of two variables across different groups, such as comparing the average income levels of men and women or comparing the incidence of a disease between different age groups.

3. #### **Testing hypotheses:** Bivariate analysis can be used to test hypotheses about the relationship between two variables, such as whether there is a significant difference in the mean values of two groups or whether there is a significant correlation between two variables.

4. #### **Predicting outcomes:** Bivariate analysis can be used to predict the value of one variable based on the value of another variable, such as predicting the price of a house based on its size or predicting the probability of an event based on the values of certain predictors.

## Scatter Plot

* #### A scatter plot is a type of chart used to visualize the relationship between two continuous variables. 

* #### It is a two-dimensional plot in which one variable is represented on the horizontal axis (x-axis) and the other variable is represented on the vertical axis (y-axis). Each point on the plot represents a pair of values for the two variables.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some sample data
temperature = np.concatenate([np.random.normal(20, 5, 25), np.random.normal(30, 5, 25), np.random.normal(40, 5, 25)])
ice_cream_sales = np.concatenate([np.random.normal(100, 30, 25), np.random.normal(200, 30, 25), np.random.normal(300, 30, 25)])

# Create a scatter plot
plt.scatter(temperature, ice_cream_sales)

# Add labels and title
plt.xlabel('Temperature (Celsius)')
plt.ylabel('Ice Cream Sales')
plt.title('Ice Cream Sales vs Temperature')

# Show the plot
plt.show()
```

## **What is a Correlation?**

#### Correlation between two variables is a statistical measure of the strength and direction of the relationship between them. Correlation is a value between -1 and 1, where -1 represents a perfect negative correlation, 0 represents no correlation, and 1 represents a perfect positive correlation.



## **Correlation in a Scatter Plot:**

#### In a scatter plot, the correlation between two variables is shown by the overall pattern of the data points. 

* #### Positive Correlation
* #### Negative Correlation
* #### No Correlation

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the data
x = np.random.rand(50)
y1 = x + np.random.rand(50)       # positive correlation
y2 = 2 - x + np.random.rand(50)   # negative correlation
y3 = np.random.rand(50)           # no correlation

# Plot the data
fig, axs = plt.subplots(1, 3, figsize=(12, 4))

# Plot scatter plot with positive correlation
axs[0].scatter(x, y1)
axs[0].set_title('Positive Correlation')
z1 = np.polyfit(x, y1, 1)
p1 = np.poly1d(z1)
axs[0].plot(x,p1(x),"r--")

# Plot scatter plot with negative correlation
axs[1].scatter(x, y2)
axs[1].set_title('Negative Correlation')
z2 = np.polyfit(x, y2, 1)
p2 = np.poly1d(z2)
axs[1].plot(x,p2(x),"r--")

# Plot scatter plot with no correlation
axs[2].scatter(x, y3)
axs[2].set_title('No Correlation')
z3 = np.polyfit(x, y3, 1)
p3 = np.poly1d(z3)
axs[2].plot(x,p3(x),"r--")

# Show the plot
plt.show()
```

## **Corelation using Pandas and Seaborn**

#### [Seaborn Pair](https://seaborn.pydata.org/generated/seaborn.pairplot.html)

```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

iris = sns.load_dataset("iris")
```

The Iris Dataset contains four features (length and width of sepals and petals) of 50 samples of three species of Iris (Iris setosa, Iris virginica and Iris versicolor).

```python
iris.sample(10)
```

#### The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column.

```python
g = sns.pairplot(iris)
```

```python
g = sns.pairplot(iris, hue="species", diag_kind='hist')
```

```python
g = sns.pairplot(iris, vars=["sepal_width", "sepal_length"])
```

```python
g = sns.pairplot(
    iris, 
    x_vars=["sepal_width", "sepal_length"], 
    y_vars=["petal_width", "petal_length"])
```

```python
g = sns.pairplot(iris, kind="reg")
```

```python
g = sns.pairplot(iris, kind="reg", hue='species')
```
